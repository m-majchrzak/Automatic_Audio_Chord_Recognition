```python
import os
from collections import defaultdict
import csv
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
chord_vocabulary = ['C:maj', 'C#:maj', 'D:maj', 'D#:maj', 'E:maj', 'F:maj', 'F#:maj', 'G:maj', 'G#:maj', 'A:maj', 'A#:maj', 'B:maj',
                        'C:min', 'C#:min', 'D:min', 'D#:min', 'E:min', 'F:min', 'F#:min', 'G:min', 'G#:min', 'A:min', 'A#:min', 'B:min',
                        'N', 'X']
```

```python
# Function to plot chord occurrences as a bar plot
def plot_chord_counts(chord_count_file, dataset_name):
    # Read the chord count file into a pandas DataFrame
    chord_data = pd.read_csv(chord_count_file)

    # Create a bar plot using seaborn
    plt.figure(figsize=(10, 6))
    sns.barplot(x='Chord', y='Count', data=chord_data, order=chord_vocabulary)
    plt.title(f'Chord Occurrences in {dataset_name} dataset')
    plt.xticks(rotation=90)
    plt.tight_layout()
    plt.show()
```

```python
# Function to plot chord progressions as a heatmap
def plot_chord_progressions(progression_file, dataset_name):
    # Read the chord progression file into a pandas DataFrame
    progression_data = pd.read_csv(progression_file, index_col=0)

    # Create a heatmap using seaborn
    plt.figure(figsize=(12, 8))
    sns.heatmap(progression_data, annot=False, cmap='Blues', cbar=True, linewidths=0.5)
    plt.title(f'Chord Progressions Heatmap for {dataset_name} dataset')
    plt.xlabel('Next Chord')
    plt.ylabel('Previous Chord')
    plt.tight_layout()
    plt.show()
```
